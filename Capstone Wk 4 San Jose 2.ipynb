```python
# instantiate a FEATURE group for the Homeless Shelters in the dataframe
housing = folium.map.FeatureGroup()
```

```python
# loop through the Affordable Housing

for lat, lng, in zip(sanjoseLL.Latitude, sanjoseLL.Longitude):
    housing.add_child(
    folium.features.CircleMarker(
    [lat, lng],
    radius=5,
    color='white',
    fill=True,
    fill_color='blue',
    fill_opacity=0.6
    ))
    
# add pop-up text
latitudes  = list(sanjoseLL.Latitude)
longitudes = list(sanjoseLL.Longitude)
labels = list(sanjoseLL.Address)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat,lng], popup=label).add_to(sanj_HH)
    
# add streets to map
sanj_HH.add_child(housing)
```
